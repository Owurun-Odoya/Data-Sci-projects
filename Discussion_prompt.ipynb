#### 10) Can you describe what a unit test is for and what some of its limitations are?
 A unit test is a type of software testing that checks individual units or components of a software application¹. The purpose of unit testing is to validate that each unit of the software performs as designed². Unit tests are typically automated tests that are written by developers⁶. 

Some of the advantages of using unit tests are:
- It helps to identify bugs early in the development cycle.
- It helps to improve the quality of the code.
- It helps to reduce the cost of fixing bugs.

However, there are some limitations to using unit tests:
- Unit tests cannot detect integration or interfacing issues between two modules.
- Unit tests cannot catch complex errors in the system ranging from multiple modules.
- Unit tests cannot test non-functional attributes like usability, scalability, the overall performance of the system, etc.
- Unit testing increases the amount of code that needs to be written.
- It is problematic when testing your user interface (UI).
- It may affect the product’s structural design and add indirection that is pointless¹.

#### 11) What are some other strategies for testing software?
  There are many different types of software tests, each with specific objectives and strategies³. Some of the common testing strategies are:
- Black box testing – Tests the functionality of the software without looking at the internal code structure.
- White box testing – Tests the internal code structure and logic of the software.
- Unit testing – Tests individual units or components of the software to ensure they are functioning as intended.
- Integration testing – Tests the integration of different components of the software to ensure they work together as a system.
- Functional testing – Tests the functional requirements of the software to ensure they are met.
- System testing – Tests the complete software system to ensure it meets the specified requirements.
- Acceptance testing – Tests the software to ensure it meets the customer’s or end-user’s expectations.
- Regression testing – Tests the software after changes or modifications have been made to ensure the changes have not introduced new defects.
- Performance testing – Tests the software to determine its performance characteristics such as speed, scalability, and stability.
- Security testing – Tests the software to identify vulnerabilities and ensure it meets security requirements¹.

##### 13) what are the advantages of distriuted version control systems compared to centralised sysytems?

Centralized version control systems (CVCS) and distributed version control systems (DVCS) are two types of version control systems used in software development.

In a centralized version control system, the source code is stored in a central repository, and developers check out files from that repository to work on them. Changes are made locally and then committed back to the central repository. Examples of centralized version control systems include Subversion (SVN) and Perforce 12.

In contrast, a distributed version control system allows developers to create local repositories on their machines, which can be used to track changes to the source code. Developers can then share changes with others by pushing their changes to a remote repository. Examples of distributed version control systems include Git, Mercurial, and Bazaar.


Distributed version control systems (DVCS) offer several advantages over centralized version control systems. Some of these advantages include:

1. **Offline work**: DVCS allows users to work productively when not connected to a network ¹.
2. **Faster common operations**: Common operations such as commits, viewing history, and reverting changes are faster for DVCS because there is no need to communicate with a central server ¹.
3. **Private work**: DVCS allows private work, so users can use their changes even for early drafts they do not want to publish ¹.
4. **Faster actions**: Performing actions other than pushing and pulling changesets is extremely fast because the tool only needs to access the hard drive, not a remote server ¹.
5. **Local committing**: Committing new changesets can be done locally without anyone else seeing them. Once you have a group of changesets ready, you can push all of them at once ¹.
6. **Offline access**: Everything but pushing and pulling can be done without an internet connection ¹.

In summary, DVCS offers more flexibility and speed than centralized version control systems.
